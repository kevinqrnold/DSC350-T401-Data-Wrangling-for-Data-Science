## Import Data

```python
#import data as a pandas data frame
df_golfers = pd.read_csv('/Users/kevin/Documents/Bellevue Data Science/DSC310-T301 Data Analysis and Visualization/PGATOURStrokesGained2010-18.csv')
#display first three rows of data
df_golfers.head(3)
```

## Explore the Data

```python
#check for missing data
df_golfers.isna().sum()
```

```python
#get top 10 golfers by Average_SG_Total
df_golfers.sort_values(by = 'Average_SG_Total', axis = 0, ascending = False).head(10)
```

```python
#get top 10 golfers by Average_SG_Total and displaying only Player_Name, Year, and Average_SG_Total
df_golfers[['Player_Name', 'Year', 'Average_SG_Total']].sort_values(by = 'Average_SG_Total', axis = 0, ascending = False).head(10)
```

```python
#find top 5 golfers by mean Average_SG_Total
df_golfers[['Player_Name', 'Average_SG_Total']].groupby('Player_Name').mean() \
.sort_values(by = 'Average_SG_Total', axis = 0, ascending = False).head(5)
```
